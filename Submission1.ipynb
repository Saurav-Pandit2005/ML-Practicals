A Pie Chart represents the part of a wgole entity . the entire pie represents 100% , and each slice represents a proportion or percantage of th whole.

```python
df['source_city'].value_counts().plot(kind='pie', autopct='%.3f')
```

```python
df['source_city'].value_counts().plot(kind='pie', autopct='%.2f')
```

3. Domut Plots (used for categorial data)
Donut plots are used to show the proportion of each category in relation to the whole.

```python
import matplotlib as plt;
```

```python
import matplotlib.pyplot as plt;
```

```python
plt.pie(
    df['source_city'].value_counts(),
    labels=df['source_city'].value_counts().index,
    autopct='%.2f%%',
    wedgeprops=dict(width=0.4)
)
```
