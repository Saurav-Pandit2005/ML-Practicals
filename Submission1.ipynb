Key Features:
Easy to use and debug
Dynamic computation graphs

Use Cases:
Deep learning research
Natural Language Processing

3. Keras

Description:
Keras is a high-level neural network library written in Python. It runs on top of TensorFlow and is known for simplicity and fast development.

Key Features:
User-friendly API
Supports multiple backends
Fast prototyping of neural networks

Use Cases:
Deep learning models
Image classification

4. OpenCV

Description:
OpenCV (Open Source Computer Vision Library) is used for image processing and computer vision tasks.

Key Features:
Real-time image processing
Supports face detection and object tracking


Use Cases:
Face recognition
Image analysis

Description:
Scikit-Learn is a Python library used for traditional machine learning algorithms.

Key Features:
Easy to learn and use

Data classification

NLTK is a Python library for working with human language data (text).

Text processing tools

Use Cases:
Text analysis
Language processing tasks

7. SpaCy

Description:

HomeWork : implement Boxploat, Scatter Plot and Line Plot with suitable data

```python
# Boxploat

plt.figure(figsize=(6,4))
df.boxplot(column='price', by='class')
plt.xlabel('Class')
plt.ylabel('Price')
plt.title('Box Plot: Price by Class')
plt.suptitle('')
plt.show()
```

```python
# Scatter Plot

plt.figure(figsize=(6,4))
plt.scatter(df['duration'], df['price'], alpha=0.5)
plt.xlabel('Duration (hours)')
plt.ylabel('Price')
plt.title('Scatter Plot: Duration vs Price')
plt.show()
```

```python
# Line Ploat

avg_price = df.groupby('days_left')['price'].mean()

plt.figure(figsize=(6,4))
plt.plot(avg_price.index, avg_price.values)
plt.xlabel('Days Left')
plt.ylabel('Average Price')
plt.title('Line Plot: Average Price vs Days Left')
plt.show()
```
